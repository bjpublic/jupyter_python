## 예제 8-13  :  함수 매개변수에 초기값이 가변 자료형일 경우 



```python
def mutable_defaults(a,b=[]) :
    b.append(a)
    return b
    
print(mutable_defaults(10))
print(mutable_defaults.__defaults__)
print(mutable_defaults(10))
print(mutable_defaults.__defaults__)
print(mutable_defaults(10))
print(mutable_defaults.__defaults__)
```

```python
def none_defaults(a,b=None) :
    if b is None :
        b = []
        
    b.append(a)
    return b

print(none_defaults(10))
print(none_defaults(10))
print(none_defaults(10))
```

```python
object_ = object()

def object_defaults(a,b=object_) :
    if isinstance(b, object) :
        b = []
        
    b.append(a)
    return b

print(none_defaults(10))
print(none_defaults(10))
print(none_defaults(10))
```

## 예제 8-14  : 매개변수 위치에 인자 매핑하기



```python
def position_args(x,y,z) :
    return x,y,z

print(position_args(10,20))
```

```python
print(position_args(10,20,30))
```

```python
def position_args(x,y,z) :
    local = locals()
    varnames = position_args.__code__.co_varnames
    x = local[varnames[0]]
    y = local[varnames[1]]
    z = local[varnames[2]]
    return x,y,z
```

```python
print(position_args(10,20,30))
```

## 예제 8-19  : 가변 키워드인자 처리

```python
def var_key_args(**kwargs) :
    print(type(kwargs), kwargs)

print(var_key_args())
print(var_key_args(a=1,b=2,c=3))
```

```python
def var_key_args(**kwargs) :
    print(type(kwargs), kwargs)
    
    for k,v in kwargs.items() :
        print("key ",k, " value ",v)
        
print(var_key_args())
print(var_key_args(a=1,b=2,c=3))
```

```python
def var_key_args(**kwargs) :
    print(type(kwargs), kwargs)
    result = 0
    
    for v in kwargs.values() :
        result += v
    return result
        
        
print(var_key_args())
print(var_key_args(a=1,b=2,c=3))
```

## 예제 8-20  : 가변 위치인자와 키워드인자 혼용 처리


```python
def var_pos_key_args(*args,**kwargs) :
    print(type(args), args)
    print(type(kwargs), kwargs)
    result = 0
    
    for v in args :
        result += v
    for v in kwargs.values() :
        result += v
    return result

print(var_pos_key_args())
print(var_pos_key_args(1,2,3,4))
print(var_pos_key_args(1,2,3,4, a= 1, b=2))
print(var_pos_key_args( a= 1, b=2))
```

## 예제 8-23  : Sequence 자료형에 대한 인자 전달 할 때  unpacking처리


```python
def add(x,y,z) :
    print(locals())
    return x+y+z

t = (1,2,3)

print(add(t))
```

```python
print(add(*t))
```

```python
s = "str"

print(add(*s))
```

```python
l = [1,2,3]

print(add(*l))
```

```python
def add(*args) :
    print(locals())
    if isinstance(locals()["args"][0], str) :
        result = ""
    else :
        result = 0
    for i in args :
        result +=i
    return result

t = (1,2,3)
print(add(*t))
s = "str"
print(add(*s))
```

## 예제 8-24  : 딕셔너리를 unpacking해서 함수 인자 전달

```python
def mul(x,y,z) :
    print(locals())
    return x*y*z

d = {'x':1,'y':2, 'z':3}

print(mul(**d))
```

```python
def mul(**kwargs) :
    print(locals())
    result = 1
    for i in kwargs.values() :
        result *= i
    return result

d = {'x':1,'y':2, 'z':3}

print(mul(**d))
```

## 예제 8-25  :  자료형 별 반환값 언패킹 처리


```python
def rtn_tuple(a,b,c) :
    return a,b,c

def rtn_list(*args) :
    return list(args)

def rtn_str(*args) :
    return "".join(args)

print(rtn_tuple(1,2,3))
print(rtn_list(*[2,3,4]))
print(rtn_str(*"str"))
```

```python
a, *b = rtn_tuple(1,2,3)
print(a,b)
```

```python
a, *b = rtn_list(*[1,2,3])
print(a,b)
```

```python
a, *b = rtn_str(*"str")
print(a,b)
```

```python
def rtn_dict(**kwargs) :
    return kwargs

a,*b = rtn_dict(a=1,b=2,c=3)
print(a,b)
```
