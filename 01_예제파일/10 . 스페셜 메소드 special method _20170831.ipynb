```python
class Dot :
    def __init__(self) :
        self.test = 1
        self.test2 = 2
        
    def __getattribute__(self, attr) :
        print(" __getattribute__ ")
        return object.__getattribute__(self,attr)
    
    def __getattr__(self, attr) :
        print(" __getattr__ ")
        try :
            self.__dict__[attr]
        except KeyError as e :
            return "Key Error " + e.args[0]
    
```

```python
d = Dot()
print(d.test)
print(d.test2)
```

```python
print(d.test3)
```

## 예제 10-2 :   내부 속성 갱신과 삭제 


```python
class Dot1 :
    def __init__(self) :
        self.test = 1
        self.test2 = 2
        
    def __setattr__(self,attr,value) :
        print(" set attr ",attr)
        self.__dict__[attr] = value
        
    def __delattr__(self,attr) :
        print(" del attr ",attr)
        del self.__dict__[attr]
        
```

```python
d = Dot1() 

d.test3 = 100
print(d.__dict__)
```

```python
del d.test3
print(d.__dict__)
```

## 예제 10-3 : 연산자 in  : 객체 내의 원소 포함여부를 확인 


```python
class IN :
    
    def __init__(self, elements) :
        self.elements = elements
        
    def __contains__(self,element) :
        print(" __contains__")
        result = False
        for i in self.elements :
            if element == i :
                result = True
                break
        return result
```

```python
s = "원소를 찾기"

i = IN(s)
print("원" in i)
```

## 예제 10-5 : Indexing 검색  


```python
class Indexing :
    def __init__(self, content) :
        self.content = content
        
    def __getitem__(self,key) :
        print(" __getitem__ call ")
        result = ""
        for i in range(len(self.content)) :
            if i == key :
                result = self.content[key]
                break
        if len(self.content) < key :
            raise IndexError(key)
        return result
    
```

```python
s = "Inexing search"

i = Indexing(s)
print(i[0])
```

```python
print(i[30])
```

```python
class Indexing1 :
    def __init__(self, content) :
        self.content = content
        
    def __getitem__(self,key) :
        print(" __getitem__ call ")
        try :
            result = self.content[key] 
        except IndexError :
            result = self.__missing__(key)
        return result
    
    def __missing__(self,key) :
        print(" __missing__ ")
        return " no key " + (key if type(key) == str else str(key))
```

```python
s = "Inexing search"

i = Indexing1(s)
print(i[0])
```

```python
print(i[30])
```

## 예제 10-6 :  Indexing 내부 갱신 및 삭제 


```python
class Indexing1 :
    def __init__(self, content) :
        self.content = content
        
    def __getitem__(self,key) :
        return self.content[key]
    
    def __setitem__(self, key, value) :
        a = [x for x in self.content]
        if a[key] == value :
            pass 
        else :
            a.insert(key,value)
        self.content = "".join(a)
        
    def __delitem__(self, key) :
        a = [x for x in self.content if self[key] != x  ]
        self.content = "".join(a)
```

```python
i = Indexing1("갱신 및 삭제")

del i[0] 
print(i.content)
```

```python
i[0] = "갱"
print(i.content)
```

```python
i[1] = "신"
print(i.content)
```

```python
class Slicing :
    def __init__(self, content) :
        self.content = content
        
    def __getitem__(self,key) :
        return self.content[key]
    
    def __setitem__(self, key, value) :
        
        start,stop,a,j = (0,len(value),[],0)  \
                        if key.stop is None else (key.start,key.stop,[x for x in self.content],0)
      
        for i in range(start, stop) :
            if key.stop is None :
                a.insert(i,value[j])
            else :
                if len(a) > i :
                    a[i] = value[j]
                else :
                    a.append(value[j])
          
            j += 1
            if len(value) == j :
                break
            
        self.content = "".join(a)
        
    def __delitem__(self, key) :
        a = [x for x in self.content  ]
        del a[key]
        self.content = "".join(a)
```

```python
s = Slicing("슬라이싱")
```

```python
s[:] = "슬라이싱 처리"

print(s.content)
```

```python
s[1:3] = "부분"
print(s.content)
```

```python
del s[1:3]
print(s.content)
```

```python
s[0:30] = "슬라이싱 범위가 넘을 경우 처리는"
print(s.content)
```

## 예제 10-7 :  할당연산자에 대한 스페셜 메소드 처리

```python
import operator as op

class Int :
    
    def __init__(self,value) :
        self.value = value
    
    def __iadd__(self,other) :
        print(" __iadd__")
        return self.value + other
    
    def __truediv__(self,other) :
        print(" __truediv__")
        return op.truediv(self.value,other)
    
    def __floordiv__(self,other) :
        print(" __floordiv__")
        return op.floordiv(self.value,other)
    
    def __itruediv__(self,other) :
        print(" __itruediv__")
        return op.itruediv(self.value,other)
    
    def __ifloordiv__(self,other) :
        print(" __ifloordiv__")
        return op.ifloordiv(self.value,other)
    
```

```python
i = Int(100) 
i += 99
print(i)
```

```python
i = Int(100) 
print(type(i))
print(i / 3)

i /= 3
print(i)
```

```python
i = Int(100) 
print(type(i))
print(i//3)
i //= 3
print(i)
```

## 예제 10-8 :  반대 연산에 대한 스페셜 메소드 처리


```python
class RInt :
    
    def __init__(self,value) :
        self.value = value
        
    def __add__(self,other) :
        print(" __add__")
        return self.value + other
    
    def __radd__(self,other) :
        print(" __radd__")
        return other + self.value
    
```

```python
i = RInt(100)
print(i + 88)

i  = i.__radd__(99)
print(i)
```

```python
i = RInt(100)

i  = 99 + i
print(i)
```

## 예제 10-8 : 호출 가능여부 확인 

```python
class CallInstance :
    def __init__(self, name) :
        self.name = name 
        
    def __call__(self) :
        print(" instance call ")
        return self.name
```

```python
print(int)
print(callable(int))

print(int.bit_length)
print(callable(int.bit_length))

print(print)
print(callable(print))
```

## 예제 10-13 :  내부 속성을 setattr 변경

```python
class Empty :
    
    def __setattr__(self,key,value) :
        print(" __setattr__ ")
        self.__dict__[key] = value


e = Empty()

d = [ { "name" : "홍길동", "age" : 33}, 
      { "name" : "문길동", "age" : 33}, 
      { "name" : "김길동", "age" : 33}, ]

e_class = []

for elem  in d :
    e = Empty()
    for k,v in elem.items() :
        setattr(e,k,v) 
    e_class.append(e)
```

```python
print(e_class[0].__dict__)
```

```python
for i in range(len(e_class)) :         
    print(getattr(e_class[i],"name"))
    print(getattr(e_class[i],"age"))
```

```python
delattr(e_class[0],"name")
```

```python
for i in range(len(e_class)) :         
    print(getattr(e_class[i],"name"))
    print(getattr(e_class[i],"age"))
```
