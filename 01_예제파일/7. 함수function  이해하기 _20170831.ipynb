## 예제 7-23 : 함수 내에 함수 정의 및 내부 실행처리


```python
def math_func1(x,y,op) :
    
    def add_() :
        return x+y
    def sub_() :
        return x-y
    def mul_() :
        return x*y
    def div_() :
        return x/y
        
    op_func = {'+': add_, "-":sub_,
          "*":mul_, "/":div_}[op]
  
    return op_func()
    
print(math_func1(5,5,'+'))
print(math_func1(5,5,'-'))
print(math_func1(5,5,'*')) 
print(math_func1(5,5,'/'))
```

## 예제 7-24 : 내부 함수 호출을 반환에서 사용하기

```python
def math_func1(x,op) :
    
    def add_(y) :
        return x+y
    def sub_(y) :
        return x-y
    def mul_(y) :
        return x*y
    def div_(y) :
        return x/y
        
    op_func = {'+': add_, "-":sub_,
          "*":mul_, "/":div_}[op]
  
    return op_func
    
```

```python
add = math_func1(5,'+')
print(add(5))
sub = math_func1(5,'-')
print(sub(5))
mul = math_func1(5,'*')
print(mul(5))
div = math_func1(5,'/')
print(div(5))
```

## 예제 7-26 : 타 모듈의 함수를 내부 함수에 지정 



```python
import operator as op

def outer_(op_code) :
    inner_ = {'+':op.add,
             '-':op.sub,
             '*':op.mul,
             '/':op.truediv}[op_code]
    return inner_

inner = outer_('/')
print(inner)
print(inner(5,5))
```

```python
import operator as op

def outer_(op_code) :
    def inner_(x,y) :
        return  {'+':op.add,
             '-':op.sub,
             '*':op.mul,
             '/':op.truediv}[op_code](x,y)
    return inner_

inner = outer_('/')
print(inner)
print(inner(5,5))
```

## 예제 7-27 :  내부 함수로 외부 함수의 매개변수 자료형 검증


```python
def outer_check(type_code) :
    def type_check(a) :
        return isinstance(a,type_code)
    
    def inner_(x,y) :
        if not type_check(x) :
            x = type_code(x)
        if not type_check(y) :
            y = type_code(y)
        return x+y
    
    return inner_
```

```python
inner = outer_check(int)
print(inner(1.1,2.2))
```

```python
inner = outer_check(float)
print(inner(1.3,2))
```

## 예제 7-39 :  약수 구하기

```python
def common_div(x) :
    s = set()
    for i in range(1,x+1) : 
        if x % i == 0 :
            s.add(i)
    return s
            
```

```python
a = int(input(" 수를 입력하세요"))

ass = common_div(a)
b = int(input(" 수를 입력하세요"))
bss = common_div(b)

print(ass)
print(bss)

print((ass & bss))
```
