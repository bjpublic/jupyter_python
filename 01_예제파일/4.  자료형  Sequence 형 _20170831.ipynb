## 예제 4-15 :  문자열 패턴 매칭하기

```python
s = dir(str)

print(type(s))

count = 1
for i in s :
    if i.startswith("__") :
        continue
    else :
        print(i, end=" ")
        
    if count % 6 == 0 :
        print()
    count += 1
```

```python
s = dir(str)

print(type(s))

count = 1
for i in s :
    if i.endswith("__") :
        print(i, end=" ")
    else :
        continue
        
    if count % 6 == 0 :
        print()
    count += 1
```

## 예제 4-35 : 깊은 복사(deepcopy)를 하는 이유

```python
import copy

l = [1,2,3,4]

ll = [l,l]
print(ll)

lc = copy.deepcopy(ll)

print(lc)
lc[0][0] = 999

print(l)
print(lc)
```

## 예제 4-37 :  리스트의 원소 추가 삭제 하기

## 예제 4-41 :  리스트 원소들 정렬하기 : sort, reverse

```python
l4 = [1,2,3,4,5,2,2,2]

# 원소들 소팅 : 올림차수
l4.sort()
print(l4)
# 원소들 소팅 : 역순으로 소팅
l4.reverse() 
print(l4)

# 소팅 기준을 주고 역순으로 소팅

l4.sort()
print(l4)

l4.sort(reverse=True)
print(l4)
```

## 예제 4-43 :  외부에 리스트 원소를 반복해서 처리하고 내부에 넣고 처리

```python
row = [3] * 3
print(row)

li = []
for _ in range(3) :
    li.append(row)
    
print(li)

li[0][0] = 99

print(li)
print(row)
```

## 예제 4-44 : 리스트를 for 문 안에서 초기화

```python
li = []
for _ in range(3) :
    row = [3] * 3
    print(id(row))
    li.append(row)
    
print(li)

li[0][0] = 99

print(li)
print(id(row), row)
```
