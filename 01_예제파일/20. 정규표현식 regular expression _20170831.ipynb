## 예제 20-9 :   반복 :  ({m,n})


```python
import re

pattern = re.compile("m{0,}ummy")

mat = pattern.search("mummy")
print(mat.group())

mat = pattern.search("mmmmmummy")
print(mat.group())

mat = pattern.search("ummy")
print(mat.group())
```

```python
import re

pattern = re.compile("m{0,3}ummy")

mat = pattern.search("mummy")
print(mat.group())

mat = pattern.search("ummy")
print(mat.group())

mat = pattern.search("mmmmmmummy")
print(mat.group())

mat = pattern.match("mmmmmmummy")

if mat is None :
    print(" mismatching ")
else :
    print(mat.group())
```

```python
import re

pattern = re.compile("m*ummy")

mat = pattern.search("mummy")
print(mat.group())

mat = pattern.search("mmmmmummy")
print(mat.group())

mat = pattern.search("ummy")
print(mat.group())
```

## 예제 20-10: 전화번호 가져오기


```python
import re

address = "경기도 성남시 분당구 031-123-1234"
pattern = re.compile("(\d+)-(\d+)-(\d+)")

mat = pattern.search(address)
print(mat.groups())
print(mat.group())
print(mat.group(1))
print(mat.group(2))
print(mat.group(3))
```

## 예제 20-11 : 그룹 이름으로 전화번호 가져오기


```python
import re

address = "경기도 성남시 분당구 031-123-1234"
pattern = re.compile("(?P<province>\d+)-(?P<local>\d+)-(?P<number>\d+)")

mat = pattern.search(address)
print(mat.groups())
print(mat.group())
print(mat.group(1))
print(mat.group(2))
print(mat.group(3))
print(mat.group("province"))
print(mat.group("local"))
print(mat.group("number"))
```

## 예제 20-13 :  금액 표시 문자 이후에 실제 금액만 가져오기


```python
import re

pattern = "(?<=\$)([0-9.]+)"
mat = re.search(pattern,"$100000.00 888888")

print(mat.group())

pattern = "([0-9.]+).*"
mat = re.search(pattern,"$100000.00 888888")
print(mat.group())
```

## 예제 20-17 : Compile, Match객체에 대한 속성과 메소드 조회하기

```python
print(re.I)
print(pattern.flags)
print(pattern.pattern)
```

```python
import re

pattern = re.compile('[a-z]+',re.M)
print(re.M)
print(pattern.flags)
print(pattern.pattern)
```

```python
import re

pattern = re.compile('[a-z]+',re.I)

mat = pattern.search("Python")

print(mat)

for i in dir(mat) :
    if not i.startswith("_") :
        print(i)
```

```python
print(mat.string)
print(mat.span())
print(mat.pos, mat.endpos)
print(mat.start(), mat.end())
```

```python
print(mat.groups())
print(mat.group())
print(mat.groupdict())
print(mat.lastgroup)
print(mat.lastindex)
```

## 예제 20-19 :  search 함수 처리 


```python
import re

value = "파이썬은 문법은 쉽지만 개념은 어렵다"

m = re.search("(개념.*)", value)
if m:
    print("search:", m.group(1))

m = re.match("(개념.*)", value)
if m:
    print("match:", m.group(1))
else :
    print(" mismatching ")
```

```python
import re

list_ = ["dog dot", "do don't", "dumb-dumb", "no match"]

for element in list_:
    m = re.search("(?P<first>d\w+)\W(?P<last>d\w+)", element)

    if m:
        print(m.groupdict())
```

## 예제 20-21  :  sub 과 subn함수  처리 



```python
import re 

text =  "사랑이라는 한 소녀가 향수를 바르고 또 한 소년이 에프터쉐이브를 바르고 만나서"

s = re.sub("한", "", text)
print(s)

print(text.replace("한", ""))
```

```python
import re 

def sequare(match) :
    num = int(match.group())
    return str(pow(num,2) )

s = re.sub("\d+", sequare, "1 2 3")
print(s)
```

```python
import re 

text =  "사랑이라는 한 소녀가 향수를 바르고 또 한 소년이 에프터쉐이브를 바르고 만나서"

s = re.subn("한", "", text,1)
print(s)

print(text.replace("한", "",1))
```
