## 예제 22-2  : JSON 함수를 이용해서 변환하기

```python
import json

list_ = ['foo', {'bar': ('baz', None, 1.0, 2)}]
a = json.dumps(list_)
print(type(a))
print(a)

b = json.loads(a)
print(type(b))
print(b)
```

```python
import json
d = {'a': 5, '6': 7}
print(json.dumps(d))

print(json.dumps(d, sort_keys=True, indent=4))
```

```python
import json

list_ = [1,2,3,{'4': 5, '6': 7}]

print(json.dumps(list_, separators=(',', ':')))

print(json.dumps(list_, sort_keys=True, indent=4))
```

## 예제 22-3  :  파이썬 클래스를 JSON으로 변환하기


```python
class User(object):
    def __init__(self, name, password):
        self.name = name
        self.password = password
        
alice = User('Alice A. Adams', 'secret')

def jdefault(o):
    return o.__dict__

print(json.dumps(alice, default=jdefault))
```

```python
import json

def jdefault(o):
    if isinstance(o, set):
        return list(o)
    return o.__dict__

pets = set(['Tiger', 'Panther', 'Toad'])
print(json.dumps(pets, default=jdefault))
```

## 예제 22-4  : JSON을 파이썬 클래스로 변환하기


```python
import json
from collections import OrderedDict

data = json.loads('{"foo":1, "bar": 2}', object_pairs_hook=OrderedDict)

print(type(data))
print(json.dumps(data, indent=4))
```

## 예제 22-5  :  임시 파일 처리  

```python
from io import StringIO
import json

io = StringIO()
json.dump(['streaming API'], io)
print(io.getvalue())

io = StringIO(io.getvalue())
print(type(io))
print(json.load(io))
```

## 예제 22-6  : 파일 처리 

```python
import json

data = {
   'name' : 'Song',
   'shares' : 100,
   'price' : 542.23
}

json_str = json.dumps(data)

print(json_str)

data = json.loads(json_str)
print(data)
```

```python
with open('data.json', 'w') as f:
    json.dump(data, f)
```

```python
with open('data.json', 'r') as f:
    data = json.load(f)
    print(data)
```

```python
import json

f = open("file_txt.json",'w+')
f.write('[{"a": "A", "c": 3.0, "b": [2, 4]}]')
f.flush()
f.seek(0)

a = json.load(f)
print(type(a))
print(a)
f.close()
```

## 예제 22-7  :  한글 처리 

```python
import pprint
import json

book = {}
book['title'] = '한글로 처리하기'
book['tags'] = ('파이썬', 'Kindle', 'Light')
book['published'] = True
book['comment_link'] = None
book['id'] = 1024

with open('ebook.json',  'w') as f:
    json.dump(book, f)
```

```python
with open('ebook.json',  'r') as f:
    book = json.load(f)
    print(json.dumps(book,indent=4))
    
pprint.pprint(book)
```
