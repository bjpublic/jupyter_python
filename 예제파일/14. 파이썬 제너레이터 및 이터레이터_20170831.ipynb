## 예제 14-4  :  사용자 클래스 반복자 처리



```python
import collections.abc as cols

class SeqIterator :
    
    def __init__(self,seq) :
        self.seq = seq
        
    def __iter__(self) :
        return (n for n in self.seq)
    
    def __next__(self) :
        return (n for n in self.seq)
```

```python
s = SeqIterator("반복자처리")
siter = iter(s)

print(siter)
print(issubclass(type(siter), cols.Iterator))
```

```python
print(next(siter))
print(next(siter))
print(next(siter))
print(next(siter))
print(next(siter))
print(next(siter))
```

## 예제 14-21  :   Apply 클래스를 이용해서 제너레이터 처리


```python
def find(l) :
    for i in range(len(l)) :
        yield l[i]
```

```python
print(find)
```

```python
print(find([1,2,3,4]))
```

```python
class Apply :
    def __init__(self,func,*args, **kwargs) :
        self.func = func
        self.args = args
        self.kwargs = kwargs
        
    def __iter__(self) :
        return self.func(*self.args,**self.kwargs)
```

```python
apply = Apply(find,[1,2,3,4])
print(apply)
a = iter(apply)
print(a)
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
```

## 예제 14-25 :  코루틴으로 평균 처리하기



```python
def mean() :
    total = 0
    average = 0
    count = 0
    while True :
        i = yield average
        total += i
        count += 1
        average = total/count
        
```

```python
mean = mean() 
print(mean)
```

```python
mean.send(None)
```

```python
print(mean.send(100))
print(mean.send(200))
```

```python
mean.close()
```

## 예제 14-27 :  코루틴을 호출해서 처리하기


```python
la = []

def cor_sub() :
    while True :
        var = yield
        la.append(var)
```

```python
def cor_main() :
    count = 0
    while True :
        var = yield
        if count == 0 :
            global cor_sub
            cor_sub = cor_sub()
            cor_sub.send(None)
            count +=1
            
        cor_sub.send(var)
```

```python
cor_main = cor_main()
cor_main.send(None)
```

```python
cor_main.send(100)
```

```python
cor_main.send(200)
```

```python
cor_main.send(300)
```
