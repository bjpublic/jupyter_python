## ì˜ˆì œ 19-18 : íŒŒì´ì¬ ê°ì²´ë¥¼pickleë¡œ ì²˜ë¦¬í•˜ê¸°



```python
import pickle
import operator as op


l = [op.add,op.sub]

with open("obj.pickle", "wb") as f :
    pickle.dump(l,f)
    
with open("obj.pickle", "rb") as f :
    a = pickle.load(f)
    print(a[0](5,6))
    print(a[1](5,4))
```

```python
import pickle
import operator as op


s = "í”¼í´ì²˜ë¦¬"
d = {'add':op.add}

with open("obj.pickle", "wb") as f :
    pickle.dump(s,f)
    pickle.dump(d,f)
    
with open("obj.pickle", "rb") as f :
    a = pickle.load(f)
    b = pickle.load(f)
    print(a)
    print(b["add"](10,10))
```

```python
import pickle
import operator as op


l = [op.add,op.sub]

s = pickle.dumps(l)
print(s)
    
a = pickle.loads(s)
print(a)
print(a[0](5,6))
print(a[1](5,4))
```
